```python
# TODO
def get_pps(fila):
    print(f"price: {fila['price']} / surf: {fila['surface']}: {float(fila['price'])/float(fila['surface'])}")
    return float(fila['price'])/float(fila['surface'])

var3 = exe4.get_group('Valdemorillo').dropna(subset=['price','surface']) #Eliminando todos las rows que contengan Na en price o surface
var4 = exe4.get_group('Galapagar').dropna(subset=['price','surface']) #Eliminando todos las rows que contengan Na en price o surface
#var3 = exe4.get_group('Valdemorillo') #Funciona, pero el resultado arroja cosas extra
#var4 = exe4.get_group('Galapagar') #Funciona, pero el resultado arroja cosas extra
var3["pps"] = var3.apply(get_pps, axis=1)#creando la columna pps
var4["pps"] = var4.apply(get_pps, axis=1)#creando la columna pps
print(f"Precio por m2 de Valdemorillos: {var3['pps'].mean()}")#Imprimiendo el promedio de precio por m2
print(f"Precio por m2 de Galapagar: {var4['pps'].mean()}")#Imprimiendo el promedio de precio por m2
```

Los promedios obtenidos muestran que el precio del m2 en Galapagar es más costoso que en Valdemorrillo. #### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
# TODO
def mean_norm(df_input,mn,stdv):
    return df_input.map(lambda x: (x - mn) / stdv)

exe18 = subcj.groupby('level5') #Dataframe agrupado
norm = []
for name, group in exe18:    #itera en cada grupo
    aux_list = [] #lista auxiliar
    mn = group['price'].mean() #Obtención de la media del grupo
    stdv = group['price'].std() #Obtención de la desviación estándar
    #print(name,mn,stdv)
    aux_list.append(name) #Se guarda el nombre del grupo en el índice 0
    aux_list.append(mean_norm(group['price'],mn,stdv)) #Guardamos la lista de normalización en la lista aux
    norm.append(aux_list) #Agregamos la lista auxiliar en norm para formato [[][]]

#Creación de los historiagramas
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows=2, ncols=2) 

colors = ['blue']
n_bins = 20

ax0.hist(norm[0][1], n_bins, density=True, histtype='bar', color=colors, label='precio normalizado')
ax0.legend(prop={'size': 10})
ax0.set_title(f"{norm[0][0]}")

ax1.hist(norm[1][1], n_bins, density=True, histtype='bar', color=colors, label='precio normalizado')
ax1.legend(prop={'size': 10})
ax1.set_title(f"{norm[1][0]}")

ax2.hist(norm[2][1], n_bins, density=True, histtype='bar', color=colors, label='precio normalizado')
ax2.legend(prop={'size': 10})
ax2.set_title(f"{norm[2][0]}")

ax3.hist(norm[3][1], n_bins, density=True, histtype='bar', color=colors, label='precio normalizado')
ax3.legend(prop={'size': 10})
ax3.set_title(f"{norm[3][0]}")

fig.tight_layout()
plt.show()
```
